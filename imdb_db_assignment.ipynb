```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

nltk.download("stopwords", quiet=True)
nltk.download("wordnet", quiet=True)
nltk.download("omw-1.4", quiet=True)

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def clean_text(text):

    text = str(text).lower()

    # Remove HTML
    text = re.sub(r'<.*?>', ' ', text)

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove numbers and special characters
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize and remove stopwords
    words = [word for word in text.split() if word not in stop_words]

    # Lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]

    # Stemming
    words = [stemmer.stem(word) for word in words]

    return " ".join(words)

df['review'] = df['review'].apply(clean_text)
```

```python
# Get all words from all reviews using basic Python
all_words = []
for review in df['review']:
    all_words.extend(review.split())

# Calculate total number of words
total_words = len(all_words)

# Calculate total number of unique words (vocabulary)
vocabulary = set(all_words)
total_unique_words = len(vocabulary)

print(f"Total number of words in the corpus: {total_words}")
print(f"Total number of unique words (vocabulary): {total_unique_words}")
```

```python
# Problem 6

# Apply tf-idf and find out the idf scores of words, also find out the vocabulary.
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizer = CountVectorizer(
    max_features=10000,
    stop_words='english',
    min_df=5,
    max_df=0.8,
    ngram_range=(1,2)
)

X = vectorizer.fit_transform(df['review'])

print("Vocabulary Size:", len(vectorizer.get_feature_names_out()))

feature_names = vectorizer.get_feature_names_out()

bow_df = pd.DataFrame(
    X[:5].toarray(),
    columns=feature_names
)

display(bow_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tfidf = TfidfVectorizer(
    max_features=10000,
    stop_words='english',
    min_df=5,
    max_df=0.8,
    ngram_range=(1,2)
)

X = tfidf.fit_transform(df['review'])

print("Vocabulary Size:", len(tfidf.get_feature_names_out()))

feature_names = tfidf.get_feature_names_out()

tfidf_df = pd.DataFrame(
    X[:5].toarray(),
    columns=feature_names
)

display(tfidf_df)
```
